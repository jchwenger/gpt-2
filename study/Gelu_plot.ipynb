## Gelu et al. plotted

See:
- the [original paper](https://arxiv.org/pdf/1606.08415.pdf)
- this [discussion on the Stack](https://datascience.stackexchange.com/questions/49522/what-is-gelu-activation) 

N.B: erf() means 'error function', cf. [here](http://mathworld.wolfram.com/Erf.html)

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
```

```python
tf.enable_eager_execution()
```

```python
x = np.arange(-5, 2, 0.1)
y = 0.5*x*(1+tf.tanh(np.sqrt(2/np.pi)*(x+0.044715*tf.pow(x, 3))))
plt.grid(b=True, which='major', color='#666666', linestyle='-')
plt.plot(x, y, label='gelu')
plt.plot(x, tf.nn.relu(x), label='relu')
plt.plot(x, tf.nn.leaky_relu(x), label='leaky relu')
plt.plot(x, tf.nn.elu(x), label='elu')
plt.legend()
```
